```python
%reset -f

import pandas as pd


mh_index = pd.read_csv('preprocessed/mh_index/mh3d_score.csv',sep=',')
municipality_list = mh_index['municipality'].unique()
# mh_index
len(municipality_list)
```

```python
income = pd.read_csv('preprocessed/preprocessed_data/income_processed.csv',sep=',')
filtered_income = income[income['municipality'].isin(municipality_list)]
# filtered_income['year'] = mh_index['Year'].astype(int)
# filtered_income = filtered_income[filtered_income['Year'] >=2008]
filtered_income_municipality_list = filtered_income['municipality'].unique()
filtered_income
# print(filtered_income[])
# del filtered, income
```

```python
bankruptcy = pd.read_csv('preprocessed/preprocessed_data/bankruptcy_processed.csv')
filtered_bankruptcy = bankruptcy[bankruptcy['municipality'].isin(filtered_income_municipality_list)]
print(len(filtered_bankruptcy['municipality'].unique()))
# del filtered, bankruptcy
```

```python
business_loan = pd.read_csv('preprocessed/preprocessed_data/business_loan_preprocessed.csv')
business_loan
filtered_business_loan = business_loan[business_loan['municipality'].isin(filtered_bankruptcy['municipality'].unique())]
print(len(filtered_business_loan['municipality'].unique()))
# del filtered,business_loan
```

```python
kela = pd.read_csv('preprocessed/preprocessed_data/kela_data_yearly_en.csv')
kela
filtered_kela = kela[kela['municipality'].isin(filtered_business_loan['municipality'].unique())]
print(len(filtered_kela['municipality'].unique()))
# del filtered,kela
```

```python
loan_housing = pd.read_csv('preprocessed/preprocessed_data/loan_housing.csv')
loan_housing
filtered_housing =loan_housing[loan_housing['municipality'].isin(filtered_kela['municipality'].unique())]
print(len(filtered_housing['municipality'].unique()))
# del filtered_housing
```

```python
other_debt = pd.read_csv('preprocessed/preprocessed_data/other_debt_preprocessed.csv')
other_debt
filtered_other =other_debt[other_debt['municipality'].isin(filtered_housing['municipality'].unique())]
print(len(filtered_other['municipality'].unique()))
# del filtered_other
```

```python
population = pd.read_csv('preprocessed/preprocessed_data/pop_density_processed.csv')
filtered_population =population[population['municipality'].isin(filtered_other['municipality'].unique())]
print(len(filtered_population['municipality'].unique()))
# del filtered_population
```

```python
poverty = pd.read_csv('preprocessed/preprocessed_data/poverty_rate.csv')
filtered_poverty =poverty[poverty['municipality'].isin(filtered_population['municipality'].unique())]
print(len(filtered_poverty['municipality'].unique()))
# del filtered_poverty
```

```python
student_loan = pd.read_csv('preprocessed/preprocessed_data/student_loan_preprocessed.csv')
filtered_student_loan =student_loan[student_loan['municipality'].isin(filtered_poverty['municipality'].unique())]
print(len(filtered_student_loan['municipality'].unique()))
# del filtered_student_loan
```

```python
mean_debt = pd.read_csv('preprocessed/preprocessed_data/total_debt_preprocessed.csv')
filtered_mean_debt =mean_debt[mean_debt['municipality'].isin(filtered_student_loan['municipality'].unique())]
print(len(filtered_mean_debt['municipality'].unique()))
# del filtered_mean_debt
```

```python
#Common municipalities are 218
common_municipality_list = filtered_mean_debt['municipality'].unique()
mh_index['year']=mh_index['year'].astype(int)
mh_index = mh_index[(mh_index['year'] >=2008) & (mh_index['year']<=2022)]
mh_index =mh_index[mh_index['municipality'].isin(common_municipality_list)]
mh_index
```

```python
bankruptcy['year']=bankruptcy['year'].astype(int)
bankruptcy = bankruptcy[(bankruptcy['year'] >=2008) & (bankruptcy['year'] <=2022)]
bankruptcy =bankruptcy[bankruptcy['municipality'].isin(common_municipality_list)]
bankruptcy
```

```python
business_loan['year']=business_loan['year'].astype(int)
business_loan = business_loan[(business_loan['year'] >=2008) & (business_loan['year'] <=2022)]
business_loan =business_loan[business_loan['municipality'].isin(common_municipality_list)]
business_loan
```

```python
income['year']=income['year'].astype(int)
income = income[(income['year'] >=2008) & (income['year'] <=2022)]
income =income[income['municipality'].isin(common_municipality_list)]
income
```

```python
kela['year']=kela['year'].astype(int)
kela = kela[(kela['year'] >=2008) & (kela['year'] <=2022)]
kela =kela[kela['municipality'].isin(common_municipality_list)]
kela
```

```python
loan_housing['year']=loan_housing['year'].astype(int)
loan_housing = loan_housing[(loan_housing['year'] >=2008) & (loan_housing['year'] <=2022)]
loan_housing =loan_housing[loan_housing['municipality'].isin(common_municipality_list)]
loan_housing
```

```python
other_debt['year']=other_debt['year'].astype(int)
other_debt = other_debt[(other_debt['year'] >=2008) & (other_debt['year'] <=2022)]
other_debt =other_debt[other_debt['municipality'].isin(common_municipality_list)]
other_debt
```

```python
population['year']=population['year'].astype(int)
population = population[(population['year'] >=2008) & (population['year'] <=2022)]
population = population[population['municipality'].isin(common_municipality_list)]
population
```

```python
poverty['year']=poverty['year'].astype(int)
poverty = poverty[(poverty['year'] >=2008) & (poverty['year'] <=2022)]
poverty =poverty[poverty['municipality'].isin(common_municipality_list)]
# poverty.columns = ['municipality','year','poverty_rate']
poverty
```

```python
student_loan['year']=student_loan['year'].astype(int)
student_loan = student_loan[(student_loan['year'] >=2008) & (student_loan['year'] <=2022)]
student_loan =student_loan[student_loan['municipality'].isin(common_municipality_list)]
student_loan
```

```python
mean_debt['year']=mean_debt['year'].astype(int)
mean_debt = mean_debt[(mean_debt['year'] >=2008) & (mean_debt['year'] <=2022)]
mean_debt = mean_debt[mean_debt['municipality'].isin(common_municipality_list)]
mean_debt
```

```python
from functools import reduce
dfs = [mh_index,bankruptcy,business_loan,income,kela,loan_housing,other_debt,population,poverty,student_loan,mean_debt]
all_data = reduce(lambda left, right: pd.merge(left, right, on=['year', 'municipality'], how='inner'), dfs)
```

```python
all_data.to_csv('merged_all_variables.csv',index=False)
```
